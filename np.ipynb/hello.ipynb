```python
import numpy as np

# Create two one-dimensional arrays
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])

# Perform element-wise addition
# This is much faster and more efficient than using Python loops
result_sum = array1 + array2

print("Array 1:", array1)
print("Array 2:", array2)
print("Sum of Arrays (element-wise):", result_sum)
```

```python
import pandas as pd

# 1. Create a dictionary of data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

# 2. Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# 3. Print the DataFrame
print(df)
```
