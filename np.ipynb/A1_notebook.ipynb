```python
# now checking for missing value isna=to find null value
missing_count = df.isna().sum()

missing_percent=(missing_count/ len(df)*100).round(2)
missing_table=(
    pd.DataFrame({
        "missing_count" :missing_count,
        "missing_percent": missing_percent
    })
    .sort_values(by="missing_percent",ascending=False)

)
missing_table[missing_table["missing_count"]>0].head(15)
```

```python
# inplace  = that function is used to make the changes in the original file permanently

df.columns
```

```python
structural_cat_features =[
'Pool QC',
       'Fence', 'Misc Feature', 'Misc Val', 'Mo Sold', 'Yr Sold', 'Sale Type',
       'Sale Condition', 'SalePrice']

df[structural_cat_features]=df[structural_cat_features].fillna("None")
```

```python
structural_num_features =[
'Garage Yr Blt',
       'Garage Area','Garage Cars']

df[structural_num_features]=df[structural_num_features].fillna(0)
```

```python
# compute median frontage per neighborhood
lot_frontage_median = (
    df.groupby("Neighborhood")["Lot Frontage"]
    .median()
)
# Apply neighborhood wise median
df["Lot Frontage"]=df.apply(
    lambda row: lot_frontage_median[row["Neighborhood"]]
    if pd.isna(row["Lot Frontage"]) else row["Lot Frontage"],axis=1
)
```
