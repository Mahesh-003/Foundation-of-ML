# GroupBy in pandas


```python
df=pd.DataFrame({
    "Name": ["Aman", "Sourav", "karan", "Shubham", "Abhishek", None, "Sameer"],
    "Age": [22,23,None,22,25,24,26],
    "marks":[85,None,88,90,56,72,None],
    "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]})

df
```

```python
df.groupby('Subject')['marks'].mean() #mean of  the marks of each subject
```

```python
df.groupby('Subject')['Name'].count() #count students per subject
```

```python
df.groupby('Subject').size() #this will include the  missing values in the count also
```

```python
#multiple aggregation
df.groupby('Subject')['marks'].agg(['mean','max','min','count'])
```

```python
#Group by multiple columns
df.groupby(['Subject','Age'])['marks'].mean()
```
