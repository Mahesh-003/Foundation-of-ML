```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0, 5, 11)
y=x**2
y
```

```python
x
```

```python
plt.plot(x,y,'r')#'r' is the colour red
plt.xlabel('X Axis Title Here')
plt.ylabel('y Axis Title Here')
plt.title('string tittle here')
plt.show()
```

```python
# plt.subplot(nrows, ncols, plot_no)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')# more on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

```python
# create figure (empty canvas)
fig=plt.figure()

# add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])#left, bottom, width, heigth (range 0 to 1)

# plot on that set of axes
axes.plot(x,y, 'b')
axes.set_xlabel('Set x lable') # notice the use of set_ to begin methods
axes.set_ylabel('Set y lable')
axes.set_title('set tittle')
```
