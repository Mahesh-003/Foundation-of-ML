```python
# ---------EXERCISE NO 2-----------------
import numpy as np
marks= np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
marks
```

```python
print(np.shape(marks)) # shapes
print(marks.size) #size and total no of element
```

```python
print(marks[1]) # marks of 2 student
print(marks[:,2])# marks of 3 subject
print(marks[0:2]) # marks of stu1 and stu2
```

```python
avg1= np.sum(marks[:,0]) 
print(avg1/5) #for marks of ist sub
avg2= np.sum((marks[:,1])/5)
print(avg2)
print(np.sum((marks[:,2])/5))
```

```python
print(np.sum((marks[0,:])/5)) # for each student
print(np.sum((marks[1,:])/5))
print(np.sum((marks[2,:])/5))
```

```python
#pass \ fail mapping

result = np.where(marks > 50, "Pass", "Fail")

print(result)
```

```python
avg_marks = np.mean(marks, axis=1)
print(avg_marks)
```

```python
grades = np.where(avg_marks >= 85, "A",
         np.where(avg_marks >= 70, "B",
         np.where(avg_marks >= 50, "C", "Fail")))

print(grades)
```

```python
# copy vs view
marks2= marks.copy()
print(marks2)

marks3= marks.view()
print(marks3)
```

```python
marks3[0,0]=0
print(marks) # original one
print(marks3)# view array
print(marks2)#copy array
```
